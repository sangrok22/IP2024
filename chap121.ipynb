```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# Load an color image in grayscale
img = cv2.imread('car-49278_1280.jpg')
```

```python
cv2.imshow('sangrok',img)  #'sangrok' = 뜨는 창의 이름
cv2.waitKey(0) #아무 키 입력을 기다림
cv2.destroyAllWindows() #창을 닫음
```

```python
cv2.imwrite('messigray.png',img) #저장은 흑백으로 됨
```

```python
img = cv2.imread('messigray.png',0)
cv2.imshow('image',img)
k = cv2.waitKey(0)
if k == 27: # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('messigray2.png',img)
    cv2.destroyAllWindows()
```

```python
img = cv2.imread('messigray.png',0)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis
plt.show()
```
