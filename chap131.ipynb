##### Chap 1.3.1 Basic Operations on Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('car-49278_1280.jpg')
```

```python
plt_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(plt_img, interpolation = 'bicubic')

plt.show()
```

```python
cv2.imshow('sangrok',img)  #'sangrok' = 뜨는 창의 이름
cv2.waitKey(0) #아무 키 입력을 기다림
cv2.destroyAllWindows() #창을 닫음
```
