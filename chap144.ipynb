```python
import  cv2
import  numpy  as  np
from  matplotlib  import  pyplot  as  plt
```

```python
img  =  cv2.imread('openCv.png')
kernel  =  np.ones((5,5),np.float32)/25
dst  =  cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original') 
plt.xticks([]),  plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging') 
plt.xticks([]),  plt.yticks([])
plt.show()
```

##### 1. Averaging

```python
img  =  cv2.imread('openCv.png')
blur3  =  cv2.blur(img,(3,3))
blur15  =  cv2.blur(img,(15,15))

plt.subplot(121),plt.imshow(blur3),plt.title('blur3') 
plt.xticks([]),  plt.yticks([])
plt.subplot(122),plt.imshow(blur15),plt.title('blur5') 
plt.xticks([]),  plt.yticks([])
plt.show()
```

```python
img  =  cv2.imread('openCv.png')
blur  =  cv2.GaussianBlur(img,(15,15),0)
plt.subplot(121),plt.imshow(blur),plt.title('blur') 
plt.xticks([]),  plt.yticks([])
plt.show()
```

```python
img  =  cv2.imread('openCv.png')
median  =  cv2.medianBlur(img,15)
plt.subplot(121),plt.imshow(median),plt.title('median') 
plt.xticks([]),  plt.yticks([])
plt.show()
```
